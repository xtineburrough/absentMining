```python
import pandas as pd

# Read data from CSV file
df = pd.read_excel("matched_output.xlsx")

# Create lists of sentences and image links
sentences = list(df['Generated Sentence'])
image_links = list(df['ImageLink'])

# Create HTML file with image gallery and search bar
html = f"""
<!DOCTYPE html>
<html>
<head>
	<meta name="viewport" content="width=device-width, initial-scale=1">
	<style>
		body {{
			font-family: Arial, sans-serif;
		}}
		h1 {{
			text-align: center;
			margin-top: 20px;
			margin-bottom: 20px;
		}}
		form {{
			text-align: center;
			margin-bottom: 20px;
		}}
		input[type=text] {{
			padding: 10px;
			font-size: 17px;
			border: 1px solid grey;
			border-radius: 5px;
			display: inline-block;
			box-sizing: border-box;
		}}
		button {{
			background-color: #4CAF50;
			color: white;
			padding: 10px 20px;
			border: none;
			border-radius: 5px;
			cursor: pointer;
			margin-left: 10px;
		}}
		button:hover {{
			background-color: #45a049;
		}}
		.img-gallery {{
			display: flex;
			flex-wrap: wrap;
			justify-content: center;
		}}
		.img-gallery a {{
			margin: 10px;
			display: flex;
			flex-direction: column;
			align-items: center;
			text-align: center;
			text-decoration: none;
			color: black;
		}}
		.img-gallery img {{
			width: 100%;
			height: auto;
			border-radius: 5px;
		}}
		.img-gallery .desc {{
			margin-top: 10px;
			font-size: 16px;
		}}
	</style>
</head>
<body>
	<h1>Image Gallery</h1>
	<form>
		<label for="search">Search:</label>
		<input type="text" id="search" name="search" placeholder="Enter a keyword...">
		<button type="submit">Submit</button>
	</form>
	<div class="img-gallery">
"""

# Create image gallery
for i, sentence in enumerate(sentences):
    html += f"""
		<a href="{image_links[i]}" target="_blank">
			<img src="{image_links[i]}" alt="{sentence}">
			<div class="desc">{sentence}</div>
		</a>
	"""

# Add closing tags to HTML
html += """
	</div>
	<script>
		// Search function
		document.querySelector('form').addEventListener('submit', function(e) {
			e.preventDefault();
			let searchValue = document.getElementById('search').value.toLowerCase();
			let images = document.querySelectorAll('.img-gallery a');
			images.forEach(function(image) {
				let sentence = image.querySelector('.desc').innerHTML.toLowerCase();
				if (sentence.includes(searchValue)) {
					image.style.display = 'flex';
				} else {
					image.style.display = 'none';
				}
			});
		});
	</script>
</body>
</html>
"""

# Write HTML to file
with open("image_gallery.html", "w") as file:
    file.write(html)
```
